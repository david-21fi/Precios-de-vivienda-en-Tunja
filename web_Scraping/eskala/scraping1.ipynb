## automatizo recoleccion

```python
for i in range(1,df_url.shape[0]):
    url = str(df_url.iloc[i].values).strip("'[]'")
    browser.get(url)
    time.sleep(random.randint(11,12))
    html = browser.page_source ## sacamos el html de la pagina
    soup = bs(html,"html")
    temp=soup.find('div',{'class':'mh-estate__section mh-estate__section--attributes'})
    l=temp.find_all('li') # obtengo todo las etiquetas que me interezan
    
    valor = []
    clave = []
    ## utilizo re para obterne todos los datos que no tengan saltos de linea tab o espacios
    patron = '[\s\n\t]+'

    for i in range(len(l)):
        a = re.sub(patron,'',l[i].text)
        a = a.split(':')
        clave.append(a[0])
        valor.append(a[1])
    df1 = pd.DataFrame([valor],columns=clave)
    df = pd.concat([df,df1],ignore_index=True)
```

```python
df.drop_duplicates(subset='Tipodepropiedad',inplace=True)
```
